Gain: Gain is the improvement in accuracy that a feature brings to the branches it is used on. It is calculated for a feature each time it is used to make a new branch in a decision tree. The gain is a measure of the importance of a feature in terms of how much it contributes to improving a model's prediction. A higher average gain for a feature across all its uses indicates it is more influential in improving the model's predictions. This is a direct measure of the impact of a feature on the model's performance.

Weight: The weight of a feature refers to how many times that particular feature is used to split data across all decision trees in the model. A high weight indicates that a feature is frequently used in making splits in trees, which suggests it is important for partitioning the data. However, frequency alone doesn't tell us about the quality of the splits or the impact on the model's accuracy.

Cover: Cover measures the number of observations (data points) affected by a feature's split in a tree. It's the sum of the second-order gradient statistics (like the number of instances) for all data points that go through branches where this feature is used.  It indicates how broad the impact of a feature is over the data set. A feature that affects more data points (has a higher cover) when it is used in splits might be considered more important than one that affects fewer data points.

Total Gain: Total gain is the sum of gains for every time a feature is used to split the data in all trees. While average gain is useful, total gain can give a better picture when considering a feature's contribution across all trees. Total gain indicates the cumulative impact of a feature on the model's performance. This can be particularly insightful when a feature is used infrequently (lower weight) but makes highly beneficial splits (high gain) when it is used.

Total Cover: Total cover is the sum of the coverage metric for every time a feature is used to split data in all trees. It's the total number of data points that go through splits where this feature is used. Similar to total gain, total cover provides a cumulative measure of the breadth of impact a feature has on the dataset within the model. A feature that often creates branches with many data points might be more crucial to the model than one with fewer but more targeted splits.
