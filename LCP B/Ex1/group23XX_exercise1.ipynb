## 2. Try different optimizers, including at least Adam, RMSprop, and Nesterov. Each one should be tested for at least five values of the learning rate η (for example, including 10-5, 10-4, 10-3). The range could be different for different algorithms. Moreover, at least five fits per model should be collected to obtain statistics with an average and standard deviation of the model’s validation accuracy. Plot all results in a single panel of a figure as a function of η.

```python
# Data
x = np.array([10**-5, 10**-4, 10**-3, 10**-2, 10**-1])
optimizers = {
    "Adam": np.array([0.4050, 0.6350, 0.7775, 0.7725, 0.5950]),
    "RMSprop": np.array([0.4238, 0.6525, 0.7700, 0.7650, 0.3338]),
    "Nesterov": np.array([0.3325, 0.3900, 0.6187, 0.7837, 0.7800]),
}

colors = {"Adam": "red", "RMSprop": "blue", "Nesterov": "green"}

# Stats
plt.figure(figsize=(10,5))
for optimizer, vals in optimizers.items():
    plt.plot(x, vals, 'o-', label=optimizer, color=colors[optimizer], linewidth=2, markersize=8)
    mean_val = np.mean(vals)
    std_val = np.std(vals)
    max_val = np.max(vals)
    print(f"Stats for {optimizer} optimizer: mean={mean_val:.4f}, std={std_val:.4f}, max={max_val:.4f}")

# Plot

plt.xlabel("Learning Rates", fontsize=14, fontweight='bold')
plt.ylabel("Validation Accuracies", fontsize=14, fontweight='bold')
plt.title("Validation Accuracies across different Learning Rates", fontsize=16, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xscale('log')
plt.xticks(x, labels=[f"{rate:.0e}" for rate in x], fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

### Discussion:

As we can see from the stats, Adam optimizator has the highest mean validation accuracy (indicating that it leads to models with better validation performance) and the lowest standard deviation (indicating more consistent performance across different learning rates). That's why we chose this optimizator for further analysis.
