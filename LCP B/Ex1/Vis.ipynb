```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([10**-5, 10**-4, 10**-3, 0.01, 0.05])
val_adam = np.array([0.4288, 0.6263, 0.7850, 0.7688, 0.7638])
#val_RMS = np.array([])
#val_Nesterov = np.array([])

plt.figure(figsize=(12,6))
plt.plot(x, val_adam, 'o-', label="ADAM", color="red")
#plt.plot(x, val_RMS, 'o-', label="RMSprop", color="blue")
#plt.plot(x, val_Nesterov, 'o-', label="Nesterov", color="green")
plt.xlabel("Learning rates")
plt.ylabel("Validation accuracies")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Data
x = np.array([10**-5, 10**-4, 10**-3, 0.01, 0.1])
val_adam = np.array([0.4288, 0.6263, 0.7850, 0.7688, 0.6])
#val_rms = np.array([])
#val_nesterov = np.array([])

# Plot
plt.figure(figsize=(12,6))
plt.plot(x, val_adam, 'o-', label="ADAM", color="red", linewidth=2, markersize=8)
#plt.plot(x, val_rms, 'o-', label="ADAM", color="blue", linewidth=2, markersize=8)
#plt.plot(x, val_nesterov, 'o-', label="ADAM", color="green", linewidth=2, markersize=8)
plt.xlabel("Learning Rates", fontsize=14, fontweight='bold')
plt.ylabel("Validation Accuracies", fontsize=14, fontweight='bold')
plt.title("Validation Accuracies across Different Learning Rates for ADAM", fontsize=16, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xscale('log')
plt.xticks(x, labels=[f"{rate:.0e}" for rate in x], fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
