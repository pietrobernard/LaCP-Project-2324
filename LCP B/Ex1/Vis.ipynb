```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
x = np.array([10**-5, 10**-4, 10**-3, 10**-2, 10**-1])

optimizers = {
    "Adam": np.array([0.4050, 0.6350, 0.7775, 0.7725, 0.5950]),
    "RMSprop": np.array([0.4238, 0.6525, 0.7700, 0.7650, 0.3338]),
    "Nesterov": np.array([0.3325, 0.3900, 0.6187, 0.7837, 0.7800]),
}
val_adam = np.array([0.4050, 0.6350, 0.7775, 0.7725, 0.5950])
val_rms = np.array([0.4238, 0.6525, 0.7700, 0.7650, 0.3338])
val_nesterov = np.array([0.3325, 0.3900, 0.6187, 0.7837, 0.7800])


colors = {"Adam": "red", "RMSprop": "blue", "Nesterov": "green"}
# Plot
plt.figure(figsize=(12,6))
plt.plot(x, val_adam, 'o-', label="Adam", color="red", linewidth=2, markersize=8)
plt.plot(x, val_rms, 'o-', label="RMSprop", color="blue", linewidth=2, markersize=8)
plt.plot(x, val_nesterov, 'o-', label="Nesterov", color="green", linewidth=2, markersize=8)
plt.xlabel("Learning Rates", fontsize=14, fontweight='bold')
plt.ylabel("Validation Accuracies", fontsize=14, fontweight='bold')
plt.title("Validation Accuracies across Different Learning Rates for ADAM", fontsize=16, fontweight='bold')
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xscale('log')
plt.xticks(x, labels=[f"{rate:.0e}" for rate in x], fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
