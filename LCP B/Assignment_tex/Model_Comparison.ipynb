```python
import numpy as np
import time
import tracemalloc
import matplotlib.pyplot as plt
from scipy.stats import randint as sp_randint
from scipy.stats import uniform, loguniform

from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
#MLP
from sklearn.neural_network import MLPClassifier
#SVC
from sklearn.svm import SVC
#Random Forest
from sklearn.ensemble import RandomForestClassifier
#XGBoost
import xgboost
from xgboost import XGBClassifier
#GBC
from sklearn.ensemble import GradientBoostingClassifier
```

```python
dname="./DATA/"
str0="_XGB_24.dat"
fnamex=dname+'x'+str0
fnamey=dname+'y'+str0
x = np.loadtxt(fnamex, delimiter=" ",dtype=float)
y = np.loadtxt(fnamey)
y = y.astype(int)
N,L = len(x), len(x[0])

N_train = int(0.75*N)
x_train,y_train = x[:N_train],y[:N_train]
x_test,y_test = x[N_train:],y[N_train:]
print(f"N={N}, N_train={N_train}, L={L}")
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
n_iter_search = 20
```

```python
models = {
    "MLP": MLPClassifier(
        hidden_layer_sizes=(50, 50),
        activation='tanh',
        solver='sgd',
        alpha=0.07177886881742097, 
        learning_rate_init=0.08686601003000463),
    "SVM": SVC(
        C=396.6023828404273, 
        kernel='rbf', 
        gamma=0.0025135566617708314),
    "RandomForest": RandomForestClassifier(
        n_estimators=148,
        max_depth=21, 
        min_samples_split=5, 
        min_samples_leaf=1),
    "GBC": GradientBoostingClassifier(
        learning_rate=0.20545439917128422,
        max_depth=11,
        min_samples_split=14,
        n_estimators=414),
    "XGBoost": XGBClassifier(
        colsample_bytree=0.7854835899772805,
        eta=0.16681984881459822,
        gamma=2.4689779818219537,
        reg_lambda=0.7881286901161428,
        max_depth=9,
        n_estimators=345,
        subsample=0.6539457134966522)
}

results = {}
average_times = []
average_time_error = []
average_accs = []
average_accs_error = []
average_mem_usages = []
average_mem_usage_error = []

for model_name, model in models.items():
    times = []
    accs = []
    mem_usages = []
    for _ in range (10):
        tracemalloc.start()
        start_time = time.time()
        model.fit(x_train_scaled, y_train)
        times.append(time.time() - start_time)
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()
        accs.append(accuracy_score(y_test, model.predict(x_test_scaled)))
        mem_usages.append(peak / 1024**2)

    avg_time = np.mean(times)
    time_std = np.std(times)
    avg_acc = np.mean(accs)
    acc_std = np.std(accs)
    avg_mem_usage = np.mean(mem_usages)
    mem_usage_std = np.std(mem_usages)

    average_times.append(avg_time)
    average_time_error.append(time_std)
    average_accs.append(avg_acc)
    average_accs_error.append(acc_std)
    average_mem_usages.append(avg_mem_usage)
    average_mem_usage_error.append(mem_usage_std)
    
    results[model_name] = {
        "Average Training Time (s)": avg_time,
        "Training Time Std Dev": time_std,
        "Average Accuracy": avg_acc,
        "Accuracy Std Dev": acc_std,
        "Average Memory Usage (MB)": avg_mem_usage,
        "Memory Usage Std Dev (MB)": mem_usage_std
    }

results
```

```python
model_names = ["MLP", "SVM", "RandomForest", "GBC", "XGBoost"]
normalized_times = average_times / np.max(average_times)
normalized_time_errors = average_time_error / np.max(average_times)
normalized_accuracies = average_accs/ np.max(average_accs)
normalized_acc_errors = average_accs_error / np.max(average_accs)
normalized_mem = average_mem_usages/ np.max(average_mem_usages)
normalized_mem_errors = average_mem_usage_error / np.max(average_mem_usages)

fig, axs = plt.subplots(1, 3, figsize=(16, 7), dpi=120)

bar_colors = ['cornflowerblue', 'lightcoral', 'mediumseagreen', 'goldenrod', 'slategray']
error_kw = {'capsize': 5, 'capthick': 2, 'elinewidth': 2, 'alpha': 0.6}

axs[0].bar(model_names, normalized_accuracies, yerr=normalized_acc_errors, color=bar_colors, error_kw=error_kw, edgecolor='black')
axs[0].set_title('Normalized Accuracy by Model', fontsize=14)
axs[0].set_xlabel('Model', fontsize=12)
axs[0].set_ylabel('Normalized Accuracy', fontsize=12)
axs[0].set_xticks(range(len(model_names)))
axs[0].set_xticklabels(model_names, rotation=45, fontsize=11)
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

axs[1].bar(model_names, normalized_times, yerr=normalized_time_errors, color=bar_colors, error_kw=error_kw, edgecolor='black')
axs[1].set_title('Normalized Training Time by Model', fontsize=14)
axs[1].set_xlabel('Model', fontsize=12)
axs[1].set_ylabel('Normalized Training Time', fontsize=12)
axs[1].set_xticks(range(len(model_names)))
axs[1].set_xticklabels(model_names, rotation=45, fontsize=11)
axs[1].grid(axis='y', linestyle='--', alpha=0.7)

axs[2].bar(model_names, normalized_mem, yerr=normalized_mem_errors, color=bar_colors, error_kw=error_kw, edgecolor='black')
axs[2].set_title('Normalized Memory Usage by Model', fontsize=14)
axs[2].set_xlabel('Model', fontsize=12)
axs[2].set_ylabel('Normalized Memory Usage', fontsize=12)
axs[2].set_xticks(range(len(model_names)))
axs[2].set_xticklabels(model_names, rotation=45, fontsize=11)
axs[2].grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```
